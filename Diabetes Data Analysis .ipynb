##### 4. Are there any outliers in the features?

```python
# Function to detect outliers using IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Check for outliers in each feature (excluding 'Outcome')
for column in df.columns[:-1]:  # Exclude 'Outcome'
    outliers = detect_outliers(df, column)
    print(f"Number of outliers in {column}: {len(outliers)}")

    if len(outliers) > 0:
        print(f"Outliers in {column}:\n", outliers[column].values)
    print('-' * 50)
```

##### 16.  Are there any significant differences in blood pressure distributions for different outcomes?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Split the data into diabetic and non-diabetic groups
diabetic = df[df['Outcome'] == 1]
non_diabetic = df[df['Outcome'] == 0]

# Plot histograms to compare the blood pressure
plt.figure(figsize=(10, 6))

plt.hist(non_diabetic['BloodPressure'], bins=15, alpha=0.5, label='Non-Diabetic', color='blue')
plt.hist(diabetic['BloodPressure'], bins=15, alpha=0.5, label='Diabetic', color='red')

plt.title('Blood Pressure: Diabetic vs Non-Diabetic Individuals')
plt.xlabel('Blood Pressure (mm Hg)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# KDE plot (smoothed density plot)
plt.figure(figsize=(10, 6))
sns.kdeplot(non_diabetic['BloodPressure'], fill=True, label='Non-Diabetic', color='blue')
sns.kdeplot(diabetic['BloodPressure'], fill=True, label='Diabetic', color='red')

plt.title('Blood Pressure Distribution: Diabetic vs Non-Diabetic Individuals')
plt.xlabel('Blood Pressure (mm Hg)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([non_diabetic['BloodPressure'], diabetic['BloodPressure']], labels=['Non-Diabetic', 'Diabetic'])
plt.title('Blood Pressure Comparison: Diabetic vs Non-Diabetic Individuals')
plt.ylabel('Blood Pressure (mm Hg)')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import ttest_ind

# Perform t-test to compare blood pressure between diabetic and non-diabetic individuals
t_stat, p_val = ttest_ind(diabetic['BloodPressure'], non_diabetic['BloodPressure'], nan_policy='omit')

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpretation
if p_val < 0.05:
    print("There is a statistically significant difference in blood pressure between diabetic and non-diabetic individuals.")
else:
    print("There is no statistically significant difference in blood pressure between diabetic and non-diabetic individuals.")
```

```python
from scipy.stats import mannwhitneyu

# Perform Mann-Whitney U test to compare blood pressure between diabetic and non-diabetic individuals
u_stat, p_val_mw = mannwhitneyu(diabetic['BloodPressure'], non_diabetic['BloodPressure'], alternative='two-sided')

print(f"U-statistic: {u_stat:.4f}")
print(f"P-value: {p_val_mw:.4f}")

# Interpretation
if p_val_mw < 0.05:
    print("There is a statistically significant difference in blood pressure between diabetic and non-diabetic individuals.")
else:
    print("There is no statistically significant difference in blood pressure between diabetic and non-diabetic individuals.")
```

```python
from scipy.stats import ttest_ind

# T-test for Skin Thickness
t_stat_skin, p_val_skin = ttest_ind(diabetic['SkinThickness'], non_diabetic['SkinThickness'], nan_policy='omit')
print(f"Skin Thickness - T-statistic: {t_stat_skin:.4f}, P-value: {p_val_skin:.4f}")

# T-test for Insulin
t_stat_insulin, p_val_insulin = ttest_ind(diabetic['Insulin'], non_diabetic['Insulin'], nan_policy='omit')
print(f"Insulin Levels - T-statistic: {t_stat_insulin:.4f}, P-value: {p_val_insulin:.4f}")
```

```python
from scipy.stats import mannwhitneyu

# Mann-Whitney U test for Skin Thickness
u_stat_skin, p_val_skin_mw = mannwhitneyu(diabetic['SkinThickness'], non_diabetic['SkinThickness'], alternative='two-sided')
print(f"Skin Thickness - U-statistic: {u_stat_skin:.4f}, P-value: {p_val_skin_mw:.4f}")

# Mann-Whitney U test for Insulin
u_stat_insulin, p_val_insulin_mw = mannwhitneyu(diabetic['Insulin'], non_diabetic['Insulin'], alternative='two-sided')
print(f"Insulin Levels - U-statistic: {u_stat_insulin:.4f}, P-value: {p_val_insulin_mw:.4f}")
```
